```python
import pandas as pd
import json
```

```python
def to_list_of_lists(s: str):
    try:
        return json.loads(s)
    except Exception as ex:
        return [[]]
```

```python

goldens = pd.read_excel('data-populated.xlsx', index_col='record_id').\
    fillna('').replace('ERROR', '')
table_cols = ['Deductions_Table', 'Earnings_Table']
goldens[table_cols] = goldens[table_cols].map(to_list_of_lists)
goldens.head()
```

```python
goldens.loc['APPELF__4401-000000009041410-44012901_&I100PYSTB_20191017172943_19800261-09041410-44012901-001_3_.pdf_0', 'Deductions_Table']
```

```python
actual = pd.read_excel('actual.xlsx', sheet_name='US Paystub', index_col='file_and_record_id').\
    drop(columns=['Unnamed: 0']).fillna('').replace('ERROR', '')
actual[table_cols] = actual[table_cols].map(to_list_of_lists)
actual.head()
```

```python
def table_accuracy(table1: list, table2: list) -> float:
    try:
        df1 = pd.DataFrame(table1)
        df2 = pd.DataFrame(table2)
    except:
        df1 = pd.DataFrame([])
        df2 = pd.DataFrame([])

    # If either DataFrame is empty, return 0
    if df1.empty or df2.empty:
        return 0.0
        
    # Get number of matching cells
    matches = 0
    total_cells = 0
    
    # Compare each cell
    for i in range(min(len(df1), len(df2))):
        for col in df1.columns:
            if col in df2.columns:
                total_cells += 1
                if df1.iloc[i][col] == df2.iloc[i][col]:
                    matches += 1
                    
    # Return accuracy as percentage
    return matches / total_cells if total_cells > 0 else 0.0

# print(table_accuracy(
#     pd.DataFrame(eval(actual.loc['APPELF__4401-000000009041410-44012901_&I100PYSTB_20191017172943_19800261-09041410-44012901-001_3_.pdf_0', 'Deductions_Table'])),
#     pd.DataFrame(eval(goldens.loc['APPELF__4401-000000009041410-44012901_&I100PYSTB_20191017172943_19800261-09041410-44012901-001_3_.pdf_0', 'Deductions_Table']))))
```

```python
accuracy = pd.DataFrame(index=goldens.index, columns=table_cols)
for i in goldens.index:
    for col in table_cols:
        accuracy.loc[i, col] = table_accuracy(actual.loc[i, col], goldens.loc[i, col])
accuracy
```

```python
for col in table_cols:
    accuracy.loc['AVERAGE', col] = accuracy[col].mean()
accuracy.loc['AVERAGE', :]
```

```python
accuracy.to_csv('results.csv')
with open('results-summary.txt', 'w') as f:
    f.write(str(accuracy.loc['AVERAGE', :]))
```
